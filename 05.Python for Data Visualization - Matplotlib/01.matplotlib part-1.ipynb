```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Two Ways to create the graph or ploat
# 1. Functional Method
# 2.Object oriented Method
```

```python
# Functional Method
```

```python
plt.plot(x,y)
# Not Working
# plt.xlabel('X Axis Title Here')
# plt.ylabel('Y Axis Title Here')
# plt.title('String Title Here')
# plt.show()
```

```python
# Create The Sub Ploat in the one canvace
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'b--')
```

```python
# OOP's Method

fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('x Axes')
axes.set_xlabel('x Axes')
axes.set_title('OO Ploat')
```

```python
fig = plt.figure()
# ['left','bottom','width','height']
axes1 = fig.add_axes([0.1,0.1,0.9,0.9])
axes1.plot(x,y)
axes1.set_title('Larger Ploat')
axes2 = fig.add_axes([0.19,0.5,0.4,0.3])
axes2.plot(y,x)
axes2.set_title('Smaller Ploat')
```

```python
# Create Ths Subploat In oo Method
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
# Here axes is the array of the ploats
# axes

axes[0].plot(x,y,'b--')
axes[0].set_title('First Plot')
axes[1].plot(y,x,'r--')
axes[1].set_title('Second Plot')

# No Overlap the plot
plt.tight_layout()
plt.show()
```

## FigSize and dpi

```python
fig,axes = plt.subplots(ncols=1,nrows=2,figsize=(8,3),dpi=200)
plt.tight_layout()

axes[0].plot(x,y,'r-')
axes[0].set_title("First Plot")

axes[1].plot(y,x,'b-')
axes[1].set_title("Second Plot")
```

```python
# Save The Above Plot in the image
fig.savefig('my_fig.png')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=1,figsize=(3,3),dpi=160)
axes.plot(x,x**2,label="square")
axes.plot(x,x**3,label="cube")
axes.plot(x,x**1,label="one")

# loc define the location of the lable and o is the best case that choose by the matplotlib
axes.legend(loc = 0)
fig.savefig('graph.jpg')
```

# Plot Apperence

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=2,linestyle='-',marker='o',markersize=10,markerfacecolor='yellow',
        markeredgewidth=3,markeredgecolor='green')
# linestyle types = -,--,-.,:,steps
```

```python
# Setting the limit of the x and y axes
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=2,linestyle='-',marker='o',markersize=10,markerfacecolor='yellow',
        markeredgewidth=3,markeredgecolor='green')
ax.set_xlim([0,3])
ax.set_ylim([0,15])
```
